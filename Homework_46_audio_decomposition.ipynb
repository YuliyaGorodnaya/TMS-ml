Разложение аудио (demucs + spleeter)

```python
AUDIO_PATH = "in/song.mp3"
OUTPUT_DIR = "out"
```

2 дорожки (вокал и другое)

```python
from demucs.separate import main as demucs_main

demucs_main(["--mp3", "--two-stems", "vocals", "-n", "mdx_extra", AUDIO_PATH])
```

4 дорожки (вокал + отдельно инструменты и шум)

```python
from spleeter.separator import Separator

separator = Separator('spleeter:4stems')

separator.separate_to_file(AUDIO_PATH, OUTPUT_DIR + '/separated/unet/')
```
