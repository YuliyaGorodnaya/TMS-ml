- U-net
- Wave-U-net
- Spleeter
- Demucs

| РһСҮРёСҒСӮРәР° РҫСӮ СҲСғРјР° | U-Net; Demucs | РҘРҫСҖРҫСҲРҫ РҫРұСғСҮР°СҺСӮСҒСҸ РҪР° denoising, СҒРҫС…СҖР°РҪСҸСҺСӮ РәР°СҮРөСҒСӮРІРҫ |
