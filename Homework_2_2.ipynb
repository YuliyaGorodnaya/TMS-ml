```python
# Скрин 2. Задача 1
# Поиск значения функции при x = 1.79

import math

def calculate_y(x) -> float:
  part_1 = math.cos(math.exp(x)) + math.log((1 + x)**2)
  part_2 = math.sqrt(math.exp(math.cos(x)) + (math.sin(math.pi * x))**2)
  part_3 = math.sqrt(1 / x) + math.cos(x**2)
  return (part_1 + part_2 + part_3)**(math.sin(x))

x = 1.79
calculate_y(x)
```

```python
# Скрин 2. Задача 2
# Поиск подстроки в строке

def search_substr(subst: str, st: str):
  if subst.lower() in st.lower():
    print("Есть контакт!")
  else:
    print("Мимо!")

search_substr("Кот", "бойкот")
search_substr("кот", "бойкот")
search_substr("кит", "бойкот")
```

```python
# Скрин 2. Задача 3
# Топ-3 самых часто встречающихся символов в строке

from collections import Counter

st = "94 3553 6565 44 232 33 44"
frequency = Counter(st.replace(" ", ""))
top_3_by_frequency = frequency.most_common(3)
print(top_3_by_frequency)
```
