```python
# Операции над матрицами
# Пример 4

class Matrix:
  def __init__(self, matrix):
    self.matrix = matrix
    self.rows = len(matrix)
    self.columns = len(matrix[0])

  def __str__(self):
    return '\n'.join(['\t'.join(map(str, row))for row in self.matrix])

  def __add__(self, matrix_to_add):
    if self.rows != matrix_to_add.rows or self.columns != matrix_to_add.columns:
      raise ValueError("Нельзя сложить матрицы разных размерностей!")
    result = [[self.matrix[i][j] + matrix_to_add.matrix[i][j]
               for j in range(self.columns)] for i in range(self.rows)]
    return Matrix(result)

  def __sub__(self, matrix_to_subtract):
    if self.rows != matrix_to_subtract.rows or self.columns != matrix_to_subtract.columns:
      raise ValueError("Нельзя вычесть матрицу другой размерности!")
    result = [[self.matrix[i][j] -matrix_to_subtract.matrix[i][j]
               for j in range(self.columns)] for i in range(self.rows)]
    return Matrix(result)

  def __mul__(self, mutrix_to_multiply):
    if self.columns != mutrix_to_multiply.rows:
      raise ValueError("Число столбцов 1-ой матрицы не совпадает с числом строк 2-ой!")
    result = [[sum(self.matrix[i][k] * mutrix_to_multiply.matrix[k][j]
                   for k in range(self.columns))
                   for j in range(mutrix_to_multiply.columns)]
                   for i in range(self.rows)]
    return Matrix(result)
```

```python
A = Matrix([[1, -2, 8],[7, 5, -2],[-2, 3, 0]])
B = Matrix([[0, 2, 4],[1, -3, -2],[-2, 1, 3]])
C = Matrix([[0, 3],[5, -4],[7, -1]])

print("A + B:\n")
print(A + B)

print("\nB - A:\n")
print(B - A)

print("\nA * C\n")
print(A * C)

print("\nA * B * C\n")
print(A * B * C)
```

```python
# Операции над матрицами
# Пример 5

import numpy as np

A = np.array([[9, 2, 4],[7, 5, -5],[6, -2, 3]])
B = np.array([[8, 2, 4],[0, 1, -3],[-2, 1, 6]])
C = np.array([[-2, 0],[4, 8],[1, -1]])

print("A + B:\n", A + B)
print("\nB - A:\n", B - A)
print("\nA * C\n", np.dot(A, C))
AB = np.dot(A, B)
print("\nA * B * C\n", np.dot(AB, C))
```
