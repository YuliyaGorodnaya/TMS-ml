```python
# Скрин 2. Задача 5
# Проверка угрозы в шахматах (враг - конь/ферзь)

def are_coordinates_correct(x, y) -> bool:
  if (1 <= x <= 8) and (1 <= y <= 8):
    return True
  else:
    return False

def check_threat_from_knight(my_x, my_y, rival_x, rival_y):
  dx = abs(my_x - rival_x)
  dy = abs(my_y - rival_y)
  return (dx, dy) in [(1, 2), (2, 1)]

def check_threat_from_queen(my_x, my_y, rival_x, rival_y):
  return (rival_x == my_x or
          rival_y == my_y or
          abs(rival_x - my_x) == abs(rival_y - my_y))

while True:
  try:
    chess_piece = input("Введите тип вражеской фигуры (конь/ферзь): ")
    if chess_piece not in ["конь", "ферзь"]:
      print("Необходимо выбрать 'конь' или 'ферзь'!")
      continue

    print("Координаты вашей фигуры:")
    my_x = int(input("Введите x: "))
    my_y = int(input("Введите y: "))
    print("Координаты фигуры врага:")
    rival_x = int(input("Введите x: "))
    rival_y = int(input("Введите y: "))

    if not (are_coordinates_correct(my_x, my_y) and
            are_coordinates_correct(rival_x, rival_y)):
            print("Координаты должны быть в диапазоне от 1 до 8!")
            continue

    if (my_x, my_y) == (rival_x, rival_y):
      print("Фигуры не могут быть на одном поле!")
      continue

    if chess_piece == "конь":
      if check_threat_from_knight(my_x, my_y, rival_x, rival_y):
        print("Конь УГРОЖАЕТ вашей фигуре!")
        break
      else:
        print("Конь НЕ УГРОЖАЕТ вашей фигуре!")
        break
    else:
      if check_threat_from_queen(my_x, my_y, rival_x, rival_y):
        print("Ферзь УГРОЖАЕТ вашей фигуре!")
        break
      else:
        print("Ферзь НЕ УГРОЖАЕТ вашей фигуре!")
        break
  except ValueError:
    print("Координаты фигур должны быть целыми числами! Повторите ввод:")
```

```python
# Скрин 2. Задача 5
# Сложная задача про шахматы (наша фигура - конь; проверка двух ходов)

def find_next_positions_for_knight(x, y):
  next_positions = [
      (x + 1, y + 2), (x + 1, y - 2),
      (x - 1, y + 2), (x - 1, y - 2),
      (x + 2, y + 1), (x + 2, y - 1),
      (x - 2, y + 1), (x - 2, y - 1)
  ]
  return [(next_x, next_y) for next_x, next_y in next_positions
          if are_coordinates_correct(next_x, next_y)]

def check_threat_now(my_x, my_y, rival_x, rival_y):
  return (rival_x, rival_y) in find_next_positions_for_knight(my_x, my_y)

def check_threat_after_move(my_x, my_y, rival_x, rival_y):
  next_positions = find_next_positions_for_knight(my_x, my_y)
  for next_x, next_y in next_positions:
    if check_threat_now(next_x, next_y, rival_x, rival_y):
      return True
  return False

while True:
  try:
    my_x = int(input("Введите координату х вашей фигуры: "))
    my_y = int(input("Введите координату у вашей фигуры: "))
    rival_x = int(input("Введите координату х вражеской фигуры: "))
    rival_y = int(input("Введите координату у вражеской фигуры: "))

    if not (are_coordinates_correct(my_x, my_y) and
            are_coordinates_correct(rival_x, rival_y)):
            print("Координаты должны быть в диапазоне от 1 до 8!")
            continue

    if (my_x, my_y) == (rival_x, rival_y):
      print("Фигуры не могут быть на одном поле!")
      continue

    if check_threat_now(my_x, my_y, rival_x, rival_y):
      print("Мой конь угрожает вражеской фигуре.")
      break
    elif check_threat_after_move(my_x, my_y, rival_x, rival_y):
      print("Через один ход мой конь будет угрожать вражеской фигуре.")
      break
    else:
      print("Никакие два хода не поставят вражескую фигуру под угрозу.")
      break

  except ValueError:
      print("Координаты фигур должны быть целыми числами! Повторите ввод:")
```
