```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("data_tms.csv")
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
sex_map = {'male': 0, 'female': 1}
df['Sex'] = list(map(lambda x: sex_map[x], df['Sex']))
```

```python
df
```

```python
df.duplicated()
```

```python
df['LastName'] = df['Name'].str.split(',').str[0]
```

```python
df
```

```python
df.drop('Name', axis=1)
```

```python
df[df["LastName"] == "Graham"]
```

```python
sns.countplot(
    x="Pclass",
    data=df,
    hue="Survived",
    palette="Set2"
)

plt.title("Выживаемость в зависимости от класса билета")
plt.xlabel("Класс билета")
plt.ylabel("Количество пассажиров")
plt.legend(title="Выжил")
plt.show()
```

Больше всего выживших в первом классе, меньше всего во втором классе. Больше всего погибших в третьем.

```python
sns.countplot(
    x="Sex",
    data=df,
    hue="Survived",
    palette="Set2"
)

plt.title("Выживаемость в зависимости от пола")
plt.xlabel("Пол")
plt.ylabel("Количество пассажиров")
plt.legend(title="Выжил")
plt.show()
```

Женщин выжило больше.

```python
plt.figure(figsize=(25, 6))

sns.countplot(
    x="Age",
    data=df,
    hue="Survived",
    palette="Set2"
)

plt.title("Выживаемость в зависимости от возраста")
plt.xlabel("Возраст")
plt.ylabel("Количество пассажиров")
plt.legend(title="Выжил")

plt.xticks(range(0, int(df["Age"].max()) + 1, 5))
plt.show()
```
