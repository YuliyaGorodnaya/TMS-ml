```python
# Скрин 1. Задача 1
# Решение квадратных уравнений

import math

def solve_quadratic_equation(a: float, b: float, c: float):
  roots = []
  D = b**2 - 4*a*c
  if D > 0:
    x1 = round((-b - math.sqrt(D)) / (2*a), 3)
    x2 = round((-b + math.sqrt(D)) / (2*a), 3)
    roots.extend([x1, x2])
  elif D == 0:
    x = round(-b / (2*a), 3)
    roots.append(x)
  return roots

def print_solution(roots: list):
  length = len(roots)
  if length == 2:
    print(f"Корни уравнения: {roots[0]}, {roots[1]}.\n")
  elif length == 1:
    print(f"Единственный корень уравнения: {roots[0]}.\n")
  else:
    print("Данное уравнение корней не имеет.\n")

while True:
  try:
    a = float(input("Введите a: "))
    1 / a                      # а не может быть равно нулю
    b = float(input("Введите b: "))
    c = float(input("Введите c: "))
    break
  except ValueError:
    print("a, b и c должны быть числами! Повторите ввод:")
  except ZeroDivisionError:
    print("Коэффициент a не может быть равен нулю! Повторите ввод:")

roots = solve_quadratic_equation(a, b, c)
print_solution(roots)
```

```python
# Скрин 1. Задача 1
# Решение уравнения, сводящегося к квадратному

print("(x^2 - 8)^2 + 4*(x^2 - 8) - 5 = 0\n")

print("Решим уравнение относительно t = x^2 - 8")
roots_t = solve_quadratic_equation(1, 4, -5)
print_solution(roots_t)

print("Решим уравнение относительно х")
if len(roots_t) == 2:
  roots_x1 = solve_quadratic_equation(1, 0, -8 - roots_t[0])
  roots_x2 = solve_quadratic_equation(1, 0, -8 - roots_t[1])
  print(f"Данное уравнение имеет 4 корня:", roots_x1, roots_x2)
elif len(roots_t) == 1:
  roots_x = solve_quadratic_equation(1, 0, -8 - roots_t[0])
  print_solution(roots_x)
else:
  print("Данное уравнение корней не имеет")
```

```python
# Скрин 1. Задача 3
# Обмен 1-го и последнего эл-ов списка местами

def change(lst: list):
  length = len(lst)
  buffer = lst[0]
  lst[0] = lst[length - 1]
  lst[length - 1] = buffer
  return lst

lst = [1, 2, 3, 4, 5, 6, 7]
change(lst)
```

```python
# Скрин 1. Задача 4
# Поиск бесполезного числа

def find_useless_element(lst):
  if len(lst) == 0:
    return "Список не содержит элементов."
  else:
    max_element = lst[0]
    for i in lst[1:]:
      if i > max_element:
        max_element = i
    return round(max_element/len(lst), 3)

lst = [74, 9, 48, 125, 11]
find_useless_element(lst)
```

```python
# Скрин 1. Задача 5
# Дополнение строк до одинаковой длины

def all_eq(lst: list[str]):
  max_length = max(len(s) for s in lst)
  for i in range(len(lst)):
    lst[i] = lst[i].ljust(max_length, "_")

lst = ["Юля", "Женя", "Тимофей"]
all_eq(lst)
lst
```
